```python
#helpers
#1 : check for string in string
def is_in(word,string):
    t=string.find(word)
    if t != -1:
        return True
    else:
        return False

#2:
#helper for lang check:
def is_heb(term):
    tt=term.replace(" ","")
    for i in range(len(tt)):
        if 'HEBREW' in unicodedata.name(tt[i]):
            return True
    return False    
     
def is_arab(term):
    for i in range(len(term)):
        if 'ARABIC' in unicodedata.name(term[i]):
            return True
    return False
    
def is_eng(term):
    if (not (is_heb(term)))and (not(is_arab(term))):
            return True
    return False
```

```python
#returns 1 if not same 0 if same
def same_ev(ev,ev2):
    t=ev
    p=ev2
    non=0
    if len(ev)!=len(ev2):
        return 1
    for i in range(len(t)):
        #print(t[i])
        #print(p[i])
        if t[i]!=p[i]:
            non=1
    return(non)        
```
